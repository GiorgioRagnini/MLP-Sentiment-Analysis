```python
import kagglehub
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
import pytorch_lightning as pl
from torchmetrics.functional import accuracy, precision, recall, auroc
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, HashingVectorizer
from sklearn.datasets import load_files
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
import numpy as np
import os
import torch
```

# Loading the Dataset from Kaggle

```python
file_path = kagglehub.dataset_download('kazanova/sentiment140')

os.listdir(file_path)
data = pd.read_csv(file_path + '/training.1600000.processed.noemoticon.csv', encoding = 'latin-1', names = ['target', 'ids', 'date', 'flag', 'user', 'text'])
data.head()
```

```python
vectorizer = CountVectorizer(binary = True)
data_copy = data.copy()
data_text = data_copy['text']
target = data_copy['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_text, target, test_size=0.2, random_state = 50)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state = 50)
```

```python
X_train_T = vectorizer.fit_transform(X_train)
X_val_T = vectorizer.transform(X_val)
X_test_T = vectorizer.transform(X_test)
```
